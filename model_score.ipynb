```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train.shape, df_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, HuberRegressor
import numpy as np

X_train, y_train = df_train.drop(columns=["price"]), df_train["price"]
X_test, y_test = df_test.drop(columns=["price"]), df_test["price"]

for (model, name) in [
  (LinearRegression(), "LinearRegression"),
  (Ridge(alpha=0.07), "Ridge"),
  (Lasso(alpha=24300), "Lasso"),
  (ElasticNet(alpha=0.74, l1_ratio=0.9), "ElasticNet"),
  (HuberRegressor(), "HuberRegressor")
]:
  model.fit(X_train, y_train)
  print(f"{name}: {model.score(X_test, y_test)}")
```
